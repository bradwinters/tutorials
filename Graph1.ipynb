```python
def readData():
    '''
    Open hardcoded file, parse data anticipataed but may change
    Load just data into a numpy array, 
    '''
    f = open('graph1pass.dat', 'r')   #Smaller dataset 
    #f = open('tes1', 'r')   #alt dataset 
    #f = open('test2', 'r')   #alt dataset 
    cnt=0   #Row count, includeing header
    n=0     #count of data after header, if applicable
    Graph=[]
    print("Loading data . . .")
    print("Many files start with header data of sorts.  Count the lines to decipher header then process the rest")
    for line in f:
        xx=line.rstrip() # get rid of cr
        if cnt==0:       # known line to contain header data
            NE=xx.split(" ")
            Nodes=int(NE[0])   
            Edges=int(NE[1])
        elif cnt > 0:  # process all the rest
            aRow=xx.split(" ")
            Node=int(aRow[0])   
            Edge=int(aRow[1])
            Graph.append((Node,Edge))
            n+=1
        else:
           print("Something is wrong, bail")

        cnt+=1
        
    return Nodes, n, Graph
```
